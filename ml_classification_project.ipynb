# Human Stress Detection

- Based on these changes during sleep, stress prediction for the following day is proposed. 
- In SayoPillow.csv, you will see the relationship between the parameters- snoring range of the user, respiration rate, body temperature, limb movement rate, blood oxygen levels, eye movement, number of hours of sleep, heart rate and Stress Levels (0- low/normal, 1 – medium low, 2- medium, 3-medium high, 4 -high) that has been generated from Literature Review. 
- For more informations, access the [link](https://www.kaggle.com/datasets/laavanya/human-stress-detection-in-and-through-sleep?select=SaYoPillow.csv) to the Kaggle's Dataset.

## Model Creation

```python
from sklearn.model_selection import StratifiedShuffleSplit
strat_shuf_split = StratifiedShuffleSplit(n_splits=1, 
                                          test_size=0.3, 
                                          random_state=42)
train_idx, test_idx = next(strat_shuf_split.split(data[feature_cols], data.stress_level))

# Create the dataframes
X_train = data.loc[train_idx, feature_cols]
y_train = data.loc[train_idx, 'stress_level']

X_test  = data.loc[test_idx, feature_cols]
y_test  = data.loc[test_idx, 'stress_level']
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

### Logistic Regression

```python
lr = Pipeline([('scaler', StandardScaler()),
               ('model', LogisticRegression(solver='liblinear'))])
lr = lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)

print(classification_report(y_test, y_pred_lr))
```

```python
def plot_conf_matrix(y_pred):
    _, ax = plt.subplots(figsize=(6,6))
    ax = sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', annot_kws={"size": 20, "weight": "bold"})  
    labels = ['False', 'True']
    ax.set_ylabel('Prediction', fontsize=12);
    ax.set_xlabel('Ground Truth', fontsize=12)
```

```python
plot_conf_matrix(y_pred_lr)
```

Logistic Regression showed an incredible 100% accuracy for all metrics.

### KNN

```python
max_k = 40
f1_scores = list()
error_rates = list() # 1-accuracy

for k in range(1, max_k):
    
    knn = Pipeline([('scaler', StandardScaler()),
                    ('model', KNeighborsClassifier(n_neighbors=k, weights='distance'))])
    knn = knn.fit(X_train, y_train)
    
    y_pred_knn = knn.predict(X_test)
    f1 = f1_score(y_pred_knn, y_test, average='weighted')
    f1_scores.append((k, round(f1_score(y_test, y_pred_knn, average='weighted'), 4)))
    error = 1-round(accuracy_score(y_test, y_pred_knn), 4)
    error_rates.append((k, error))
    
f1_results = pd.DataFrame(f1_scores, columns=['K', 'F1 Score'])
error_results = pd.DataFrame(error_rates, columns=['K', 'Error Rate'])
```

```python
# Plot F1 results
sns.set_context('talk')
sns.set_style('ticks')

plt.figure(dpi=300)
ax = f1_results.set_index('K').plot(figsize=(8, 8), linewidth=6)
ax.set(xlabel='K', ylabel='F1 Score')
ax.set_xticks(range(1, max_k, 2));
plt.title('KNN F1 Score')
plt.savefig('knn_f1.png')
```

As the graph above shows, all k values gives an 100% f1-score. Therefore, let's choose a value of k=3 and use this for evaluation. 

```python
knn = Pipeline([('scaler', StandardScaler()),
                ('model', KNeighborsClassifier(n_neighbors=3, weights='distance'))])
knn = knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
y_prob_knn = knn.predict_proba(X_test)

print(classification_report(y_test, y_pred_knn))
```

```python
plot_conf_matrix(y_pred_knn)
```
